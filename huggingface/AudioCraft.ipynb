```python
from transformers import AutoProcessor, MusicgenForConditionalGeneration

processor = AutoProcessor.from_pretrained("facebook/musicgen-small")
model = MusicgenForConditionalGeneration.from_pretrained("facebook/musicgen-small")

inputs = processor(
    text=["Rap track with bassy drums and synth"],
    padding=True,
    return_tensors="pt",
)
audio_values = model.generate(**inputs, do_sample=True, guidance_scale=3, max_new_tokens=256)
```

```python
from IPython.display import Audio
import numpy as np

def play_audio_waveform(audio_waveform, sample_rate):
    # Convert PyTorch tensor to a NumPy array
    audio_waveform_np = audio_waveform.numpy()

    # Ensure the audio waveform is in the correct data type (float32)
    audio_waveform_np = audio_waveform_np.astype(np.float32)

    # Flatten the waveform if necessary (remove the extra dimensions)
    audio_waveform_np = audio_waveform_np.squeeze()

    # Play the audio using IPython.display.Audio
    return Audio(audio_waveform_np, rate=sample_rate)

sample_rate = 44100

# Play the audio
play_audio_waveform(audio_values, sample_rate)
```
